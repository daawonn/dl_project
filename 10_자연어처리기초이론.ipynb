# 딥러닝 발전 분야

- 이미지 분류
  - 딥러닝으로 가장 빨리 사용되어서 성과를 낸 분야
  - 이미지 인식대회, ResNet은 150층까지 설계해서 구현
  - 대회가 없는 이유 -> 사람의 인지능력을 추월했지만 한계도 존재 => 아직도 어려운 분야
  - 트렌드가 이미지분류에서 이미지 생성으로 옮겨갔다
  - CNN을 기반으로 작업 -> 생성을 위해서 GAN(적대적 신경망)으로 많이 사용된다 
  
- 음성 인식
  - 딥러닝을 적용하여 상용화까지 성공한 분야
  - 과거
    - GMM(음성인식) + HMM(시퀀셜정보알고리즘) + AM(HMM을 통과한 데이터를 학습해서 만든 음향모델) + LM(n-gram을 기반한 언어모델) + WFST(최종결합)을 이용해서 구현 => ASR(자동음성인식) = 복잡하다
  - 현재
    -  DNN(딥러닝 대체) -> 음향모델은 시계열 데이터를 처리하는 LSTM으로 대체, end-to-end(딥러닝의 특징) 방식 적용

## 기계 번역
  - 자연어 처리의 최종단계, 궁극의 결정체
  - 딥러닝으로 인해 정복되었다( 단, 제품별로 성능이 상이하다)
  - History
    - 1950 : RBMT (규칙기반기계번역)
    - 1990 : SMT  (통계기반 기계번역 (단어기반,구문기반)
    - 2015 : 신경망

## 생성 모델 학습 (최근 2~3년 내에 가장 유행하는 분야, 최신트렌드)
  - 이미지 분류, 텍스트분류, 판독 -> 판별모델 학습
  - 현재는 생성모델 학습에 집중
    - 데이터가 있다
    - X 에 대한 Y가 성별(남,여),기타.. 인지 찾는것이 관심사가 아니라
    - X 의 분포에 집중해서 X 자체를 묘사하는 방향으로 진행 -> 적대적학습(GAN), 변분오토인코더(VAE)등이 발전하고 있다.

# 자연어 처리 History (딥러닝 시각)

## 초기 모델 : RNN
- RNN 모델 적용
- 기존 모델 n-gram 모델 +
## word2vec
- 2013 구글 토마스 미코로프
- 단순 신경망을 이용하여 단어들을 잠재공간에 투사시켰다
  - 비슷한 단어일수록 저차원의 잠재공간에 가깝게 위치한다
  - https://projector.tensorflow.org/

## CNN
- 2014, 윤김
- 고정관념을 깼다 ( 패러다임의 변화를 유도)
- 고정관념
  - 이미지는 CNN, 텍스트는 RNN이라는 고정적인 방법론을 깨고
  - CNN을 텍스트 분류모형에 적용하여 더 좋은 성능을 냈다
  - 형태소 분석, 문장파싱, 개체명인식 등등의 해결책으로 제시
  - 텍스트CNN.png

## seq2seq, attention
- 2014
  - seq2seq 기법 개발
  - attention 기법 개발
  - => 자연어 생성기법 NLG(National Language Generation), 기계번역, 요약, 챗봇등이 등장

## memory
- 연속적인 방법으로 정보를 읽고 쓰는 기법에 초첨이 맞춰져있다
- 메모리 증가 신경망(MANN) => QA에 적합
  - 원하는 정보를 신경망을 통해서 저장하고 필요할때 조합해서 모델을 생성하고 제공한다

## 강화학습
- 영상처리분야
  - GAN : MSE의 손실함수의 한계치를 극복하기 위해 등장 -> 생성모델학습에 트렌드를 이끄는 신경망, VAE
- 자연어처리분야
  - seqGAN, 손실함수값의 한계적인 평가방법에 대한 극복차원에서 등장
  - 자연어 처리분야에서 강화학습을 사용하면 문장을 사람처럼 만들 수 있다
- 게임분야
